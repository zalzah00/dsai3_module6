```python
import pandas as pd
```

```python
df = pd.read_csv("/content/DBS_SingDollar.csv")
```

```python
df
```

```python
X = df.loc[:,["SGD"]]
Y = df.loc[:,["DBS"]]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
pred = model.predict(X)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = mean_squared_error(Y,pred) ** 0.5
```

```python
rmse
```

```python
import joblib
```

```python
joblib.dump(model,"DBS_SGD_model.pkl")
```
